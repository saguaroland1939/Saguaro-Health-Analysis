```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
csv_path = "85008_history.csv"

# Read Purchasing File and store into Pandas data frame
weather_data = pd.read_csv(csv_path)
weather_data.head()
```

```python
# List all the columns in the table
weather_data.columns
```

```python
# We only want Date, Minimum Temperature and Percipitation, so create a new table that takes that information only
new_data = weather_data[["Date time", "Maximum Temperature", "Minimum Temperature", "Temperature", "Precipitation"
                        ]]
new_data
```

```python
hot_nights = new_data.loc[new_data["Minimum Temperature"] > 89 , ["Date time",
    "Minimum Temperature", "Precipitation"]]
hot_nights
```

```python
hot_nights.count()
```

```python
# Create a scatter plot which compares over 90 and Dates
hot_nights.plot(kind="scatter", x="Minimum Temperature", y="Date time", grid=False, figsize=(10,50),
              title="Cactus Not Breathing")
plt.show()
```
