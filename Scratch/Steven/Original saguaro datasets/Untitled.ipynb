```python
import os 

import pandas as pd 

import numpy as py 
```

```python
living_data = "Current inventory and measurement data for currently living saguaros_SR.csv"

deceased_data = "Final inventory and measurement data for dead saguaros before they died_SR.csv"

full_data = "Full history of inventory and measurement data for all saguaros alive and dead_SR.csv"

```

```python
living_df = pd.read_csv(living_data,encoding='cp1252') 

living_df = pd.DataFrame(living_df)

living_df
```

```python
dec_df = pd.read_csv(deceased_data,encoding='cp1252') 

dec_df = pd.DataFrame(dec_df)

dec_df
```

```python
full_df = pd.read_csv(full_data,encoding='cp1252')

full_df
```

```python
Measure_df  = full_df.groupby(['Measurement Date','Height','Height Unit',
                               'DBH','DBH Unit','Measurement Note'], as_index=False).count()

```

```python
Measure_df = pd.DataFrame(Measure_df)

Measure_df
```

```python
Measure_df = Measure_df.drop(columns=['Accession Planting','Inventory Date',
                                      'Status','Condition','Inventory Note'
                       ,'DBH','DBH Unit'])
```

```python
Measure_df = pd.DataFrame(Measure_df)

Measure_df
```

```python
Measure_df[['Measurement Year','Time']] = Measure_df["Measurement Date"].str.split(" ",expand=True)

Measure_df
```

```python
Measure_df = Measure_df.drop(columns=['Measurement Date','Time'])

```

```python
Measure_df['Measurement Year'] = pd.to_datetime(Measure_df['Measurement Year'],errors='coerce').dt.year
```

```python
Measure_df = Measure_df.rename(columns={'Height Unit':'HeightMetric'})
```
